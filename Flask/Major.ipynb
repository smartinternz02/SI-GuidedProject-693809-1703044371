```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split

# Assuming x_train, y_train, x_test, y_test are your training and testing data
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and train a Decision Tree classifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)

# Obtain predictions for train and test sets
y_train_pred = dt.predict(x_train)
y_test_pred = dt.predict(x_test)

# Calculate metrics for train and test sets
train_accuracy = accuracy_score(y_train, y_train_pred)
train_precision = precision_score(y_train, y_train_pred, average='weighted')
train_recall = recall_score(y_train, y_train_pred, average='weighted')
train_f1score = f1_score(y_train, y_train_pred, average='weighted')

test_accuracy = accuracy_score(y_test, y_test_pred)
test_precision = precision_score(y_test, y_test_pred, average='weighted')
test_recall = recall_score(y_test, y_test_pred, average='weighted')
test_f1score = f1_score(y_test, y_test_pred, average='weighted')

# Print the metrics
print("dt-Train Accuracy:", train_accuracy)
print("dt-Test Accuracy:", test_accuracy)

print("dt-Train Precision:", train_precision)
print("dt-Test Precision:", test_precision)

print("dt-Train Recall:", train_recall)
print("dt-Test Recall:", test_recall)

print("dt-Train F1-score:", train_f1score)
print("dt-Test F1-score:", test_f1score)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier  # Import XGBoost

# Assuming your DataFrame is named df
# Assuming 'target_variable' is the name of your target variable

# Separate features (x) and target variable (y)
x = df.drop('happy', axis=1)
y = df['happy']

# Create a StandardScaler
scaler = StandardScaler()

# Fit the scaler on the features
scaler.fit(x)

# Transform the features using the fitted scaler
x_scaled = scaler.transform(x)

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)

# Create a RandomForestClassifier
rf = RandomForestClassifier()

# Train the RandomForestClassifier on the training set
rf.fit(x_train, y_train)

# Obtain predictions for the training and test sets using RandomForestClassifier
y_train_pred_rf = rf.predict(x_train)
y_test_pred_rf = rf.predict(x_test)

# Now, create an XGBoost classifier
xgb = XGBClassifier()

# Train the XGBoost classifier on the training set
xgb.fit(x_train, y_train)

# Obtain predictions for the training and test sets using XGBoost
y_train_pred_xgb = xgb.predict(x_train)
y_test_pred_xgb = xgb.predict(x_test)

# Now you can use the scaler to transform new data before making predictions
# For example:
new_data = np.array([[3, 2, 3, 5, 4, 3]])  # Replace this with your new data
scaled_new_data = scaler.transform(new_data)

# Now use the XGBoost classifier to make predictions on the scaled new data
prediction_xgb = xgb.predict(scaled_new_data)

print("RandomForestClassifier Predictions:")
print("Train:", y_train_pred_rf)
print("Test:", y_test_pred_rf)

print("XGBoost Predictions:")
print("Train:", y_train_pred_xgb)
print("Test:", y_test_pred_xgb)

print("XGBoost Prediction for New Data:", prediction_xgb)
```
